Image Processing

```python
img=cv.imread('/home/sreenal/AI/Opencv/OCR/ocr.jpeg')
img=cv.cvtColor(img,cv.COLOR_BGR2RGB)
# Grey image
def get_greyscale(image): 
    return cv.cvtColor(image,cv.COLOR_BGR2GRAY) 
    # Noise removal image 
def remove_noise(image): 
    return cv.medianBlur(image,5) 
        # Threshold image 
def threshold(image): 
    return cv.threshold(image,0,255,cv.THRESH_BINARY)[1] 


```

```python
gray = get_greyscale(img) 
```

```python
thresh = threshold(gray)
```

```python
noise= remove_noise(gray)
```

```python
plt.subplot(231) 
plt.title('gray')
plt.imshow(gray) 
plt.subplot(232) 
plt.title('noise') 
plt.imshow(noise) 
plt.subplot(233) 
plt.title('thresh')
plt.imshow(thresh) 
```
