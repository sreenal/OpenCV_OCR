```python
import cv2 as cv
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# simple brightness and contrast controls
new_image=np.zeros(img.shape,img.dtype)
alpha=1 # simple contrast control
beta=1 # Brightness control
for y in range(img.shape[0]):
    for x in range(img.shape[1]):
        for c in range(img.shape[2]):
            new_image[y,x,c]=np.clip(alpha*img[y,x,c]+beta,0,255)


# show image
cv.namedWindow('New Image',cv.WINDOW_AUTOSIZE)
cv.imshow('New Image',new_image)
cv.waitKey(0)
cv.destroyAllWindows()

```

```python
# To plot all bands(channels ) in an single plot
fig, axs = plt.subplots(nrows=1, ncols=3,figsize=(20,20)) 
for i in range(0,3): 
    ax = axs[i] 
    ax.imshow(img[:,:,i],cmap='gray') 
plt.show() 
```

```python
# Shape within the image
img_copy=img.copy()
cv.rectangle(img_copy,pt1=(20,25),pt2=(88,50),color=(255,0,0),thickness=1)
# Text within image
cv.putText(img_copy,text='Enrique',org=(28,48),fontFace=cv.FONT_HERSHEY_COMPLEX,fontScale=0.375,color=(255,255,255),thickness=1) 
plt.imshow(img_copy) 


```

Image PreProcessing

```python
img1=cv.imread('/home/sreenal/AI/Opencv/data/newspaper.webp')
```

```python
cv.namedWindow('newp',cv.WINDOW_AUTOSIZE)
cv.imshow('newp',img1)
cv.waitKey(0)
cv.destroyAllWindows()
```

```python
kernals = [3,5,11,17] 
fig, axs = plt.subplots(nrows=1,ncols=4,figsize=(20,20)) 
for ind, s in enumerate(kernals): 
    img_blurred = cv.blur(img,ksize=(s,s))
    ax = axs[ind] 
    ax.imshow(img_blurred) 
    ax.axis('off') 
plt.show() 
```

Types of Blur

```python
img_0 = cv.blur(img1,ksize=(7,7)) 
img_1 = cv.GaussianBlur(img1,ksize=(7,7),sigmaX=0) 
img_2 = cv.medianBlur(img1,7) 
img_3 = cv.bilateralFilter(img1,7,sigmaSpace=75,sigmaColor=75) 
images=[img_0,img_1,img_2,img_3] 
fig, axs = plt.subplots(nrows=1,ncols=4, figsize=(20,20)) 
for ind, p in enumerate(images): 
    ax=axs[ind] 
    ax.imshow(p) 
    ax.axis('off') 
plt.show() 
```

Thresholding

```python
_,thresh_0 = cv.threshold(img1,127,255,cv.THRESH_BINARY) 
_,thresh_1 = cv.threshold(img1,127,255,cv.THRESH_BINARY_INV) 
_,thresh_2 = cv.threshold(img1,127,255,cv.THRESH_TOZERO) 
_,thresh_3 = cv.threshold(img1,127,255,cv.THRESH_TOZERO_INV) 
_,thresh_4 = cv.threshold(img1,127,255,cv.THRESH_TRUNC) 
images=[img1,thresh_0,thresh_1,thresh_2,thresh_3,thresh_4] 
fig, axs = plt.subplots(nrows=2,ncols=3, figsize=(13,13)) 
for ind, p in enumerate(images): 
    ax=axs[ind//3,ind%3] 
    ax.imshow(p) 
    ax.axis('off') 
plt.show() 
```
